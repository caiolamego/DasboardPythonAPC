```python

import pandas as pd
import plotly.express as px
import numpy as np
from pyngrok import ngrok

df = pd.read_csv("salaries.csv") 
df.head(1333)
```

```python
df_agrupado_pelo_tamanho = df.groupby('experience_level').agg({ 'salary_in_usd': np.mean}).reset_index()

df_agrupado_pelo_tamanho
```

```python
px.bar(df_agrupado_pelo_tamanho, x='salary_in_usd', y='company_size', orientation='h')
```

```python
df_agrupado_pelo_tamanho_e_tipo  = df.groupby(['company_size' , 'employment_type']).agg({ 'salary_in_usd': np.mean}).reset_index()

df_agrupado_pelo_tamanho_e_tipo
```

```python
px.bar(df_agrupado_pelo_tamanho_e_tipo, x='salary_in_usd', y='company_size', color ='employment_type', barmode ='group', orientation='h', labels = {
    'salary_in_usd' : 'Média Salarial em Dólar' , 
    'company_size' : 'Tamanho da Empresa'
}, color_discrete_sequence=  px.colors.qualitative.Prism , template= 'plotly_dark')
```

```python
colors = px.colors.qualitative.swatches()
colors.show()
```
