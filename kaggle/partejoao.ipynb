```python
import pandas as pd
import plotly.express as px
df = pd.read_csv('/content/drive/MyDrive/salaries.csv')
df.head()
```

```python
sal_tot1= sal_tot2 = sal_tot3 = cont1 = cont2 = cont3 = 0
dados = df.values.tolist()

for linha in dados:
  if linha[8] == 100:
    cont1 += 1
    sal_tot1 += linha[6]
  elif linha[8] == 50:
    cont2 += 1
    sal_tot2 += linha[6]
  else:
    cont3 += 1
    sal_tot3 += linha[6]

media_prese = (f'Trabalho Presencial: US$ {sal_tot1/cont1:.2f}')
media_hibri = (f'Trabalho Hibrido: US$ {sal_tot2/cont2:.2f}')
media_remo = (f'Trabalho Remoto: US$ {sal_tot3/cont3:.2f}')

trab_y = [cont1, cont2, cont3]
media_x = [media_prese, media_hibri, media_remo]

fig = px.pie(names= media_x, values= trab_y)
fig.update_traces(title_text= '''Quantidade de profissionais que trablham de forma: Presencial, Remota e Híbrida
 e o Salário Médio respectivo''', title_font_size= 20) 
fig.update_traces(legendgrouptitle_text='Valor do Salário Médio em dólar:', legendgrouptitle_font_size= 13)
fig.show()
```
